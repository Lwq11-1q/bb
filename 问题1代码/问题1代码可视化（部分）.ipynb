```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 读取CSV文件
df1 = pd.read_csv('问题1答案文件/情况一第1年第1季.csv', index_col=0)
df2 = pd.read_csv('问题1答案文件/情况一第1年第2季.csv', index_col=0)
df1
```

```python
# 1. 各作物总种植面积对比（第一季vs第二季）
crop_areas1 = df1.sum()
crop_areas2 = df2.sum()
crop_areas = pd.DataFrame({'第一季': crop_areas1, '第二季': crop_areas2})
crop_areas = crop_areas[crop_areas.sum(axis=1) > 0].sort_values('第一季', ascending=False)

plt.figure(figsize=(15, 8))
crop_areas.plot(kind='bar', width=0.8)
plt.title('问题1_各作物种植面积对比（第一季vs第二季）')
plt.xlabel('作物')
plt.ylabel('总面积 (亩)')
plt.legend(title='季节')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('问题1_各作物种植面积对比.png', dpi=300, bbox_inches='tight')
plt.show()


print("所有图表已生成并保存。")
```
